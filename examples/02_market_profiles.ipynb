# 02 - Market Profiles

This notebook shows how HealthSynth can generate different commercial analytics environments using YAML profiles.

We compare:

- Canada profile
- Oncology training profile

No Python simulation code changes.
Only configuration changes.

```python
from healthsynth.generator import generate
import pandas as pd
import matplotlib.pyplot as plt
```

```python
canada = generate(
    config_path="../configs/profiles/canada.yaml",
    output_dir="../output/profile_canada",
)

oncology = generate(
    config_path="../configs/profiles/oncology_training.yaml",
    output_dir="../output/profile_oncology",
)
```

```python
def summarize_profile(name, datasets):
    return {
        "profile": name,
        "hcps": len(datasets["hcp_master"]),
        "products": len(datasets["product"]),
        "activities": len(datasets["call_activity"]),
        "prescriptions": len(datasets["prescriptions"]),
    }

pd.DataFrame([
    summarize_profile("Canada", canada),
    summarize_profile("Oncology Training", oncology),
])
```

```python
canada_specialty = canada["hcp_master"]["specialty"].value_counts(normalize=True).reset_index()
canada_specialty.columns = ["specialty", "share"]
canada_specialty["profile"] = "Canada"

oncology_specialty = oncology["hcp_master"]["specialty"].value_counts(normalize=True).reset_index()
oncology_specialty.columns = ["specialty", "share"]
oncology_specialty["profile"] = "Oncology Training"

specialty_compare = pd.concat([canada_specialty, oncology_specialty])
specialty_compare
```

```python
oncology["hcp_master"]["specialty"].value_counts(normalize=True).plot(kind="bar", figsize=(9, 5))
plt.title("Oncology Training Profile - Specialty Distribution")
plt.xlabel("Specialty")
plt.ylabel("Share of HCPs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
canada["call_activity"]["channel"].value_counts(normalize=True).plot(kind="bar", figsize=(8, 5))
plt.title("Canada Profile - Channel Mix")
plt.xlabel("Channel")
plt.ylabel("Share of Activity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
oncology["call_activity"]["channel"].value_counts(normalize=True).plot(kind="bar", figsize=(8, 5))
plt.title("Oncology Training Profile - Channel Mix")
plt.xlabel("Channel")
plt.ylabel("Share of Activity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
oncology["call_activity"]["channel"].value_counts(normalize=True).plot(kind="bar", figsize=(8, 5))
plt.title("Oncology Training Profile - Channel Mix")
plt.xlabel("Channel")
plt.ylabel("Share of Activity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Key Takeaway

HealthSynth profiles allow different commercial environments to be generated from the same engine.

This makes it possible to create training environments for different markets, therapeutic areas, or business scenarios without changing Python code.
