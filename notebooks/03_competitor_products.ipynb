# 03 – Understanding the Product Portfolio

## Why doesn't every pharmaceutical product compete in exactly the same way?

In the previous notebook we learned that market profiles define the commercial world.

One of the most important parts of that world is the product portfolio.

Products are far more than names in a table.

Each product has commercial characteristics that influence how it behaves during the simulation.

Examples include:

- manufacturer
- therapeutic area
- launch date
- baseline market share
- brand type
- commercial events
- specialty affinity

These attributes eventually influence market share, demand allocation and prescriptions.

In this notebook we'll explore the product portfolio generated by HealthSynth and understand how products differ from one another.

## The Commercial Perspective

Imagine entering a pharmaceutical market.

You might see several products treating the same disease.

Although they compete for the same patients, they rarely compete equally.

Some products are market leaders.

Some are recently launched.

Some are losing exclusivity.

Some have stronger market access.

Some are niche brands.

HealthSynth models these commercial differences rather than treating every product identically.

````python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

from healthsynth.generator import generate
````

````python
datasets = generate(
    config_path="../configs/profiles/oncology_training.yaml",
    hcps=500,
    output_dir="../output/notebook_03_products",
)
````

## Meet the Product Portfolio

Let's begin by examining the products created by the simulation.

````python
products = datasets["product"]

products
````

````python
important_columns = [
    c
    for c in [
        "product_id",
        "product_name",
        "manufacturer",
        "therapeutic_area",
        "brand_type",
        "launch_date",
        "baseline_market_share",
    ]
    if c in products.columns
]

products[important_columns]
````

## Understanding the Product Attributes

Each column contributes to the commercial simulation.

| Column | Meaning |
|---------|---------|
| Product ID | Unique identifier |
| Product Name | Brand name |
| Manufacturer | Owning company |
| Therapeutic Area | Disease area |
| Launch Date | When commercial activity begins |
| Brand Type | Originator or competitor |
| Baseline Market Share | Initial competitive assumption |

These are not merely descriptive fields.

They influence downstream commercial behaviour.

````python
plt.figure(figsize=(8,5))

plt.bar(
    products["product_name"],
    products["baseline_market_share"],
)

plt.title("Configured Baseline Market Share")

plt.ylabel("Share")

plt.tight_layout()

plt.show()
````

### Interpretation

Baseline market share represents the starting competitive position of each product.

It is not a guarantee of future performance.

Promotion, launches, market access and loss of exclusivity may all change the observed market share during the simulation.

## Product Life Cycle

Real pharmaceutical products change over time.

HealthSynth models several commercial events that influence product performance.

Examples include:

- New Product Launch
- Competitor Launch
- Loss of Exclusivity (LOE)
- Market Access Improvement

Although we will study each event separately, they all begin with product-level configuration.

```text
Discovery
     ↓
Launch
     ↓
Growth
     ↓
Maturity
     ↓
Loss of Exclusivity
     ↓
Generic Competition
```

## Products Are Connected to Other Datasets

A product appears throughout the generated market.

```text
Product
     ↓
Promotion
     ↓
Market Share
     ↓
Prescriptions
     ↓
Commercial Analytics
```

Because products are shared across multiple datasets, changes to a product's configuration can influence many downstream results.

````python
product_id = products.iloc[0]["product_id"]

print(product_id)

for dataset in [
    "call_activity",
    "promotion_effect",
    "market_share",
    "prescriptions",
]:
    print(
        dataset,
        (
            datasets[dataset]["product_id"]
            == product_id
        ).sum(),
    )
````

Rather than generating independent tables, HealthSynth treats products as shared business entities.

The same product influences:

- promotional activity
- market share
- prescriptions
- future commercial events

This consistency is one reason why the generated datasets behave like a coherent commercial system rather than unrelated synthetic tables.

## Key Takeaways

Products in HealthSynth are commercial entities rather than simple lookup records.

In this notebook we learned:

- products have meaningful commercial attributes
- product configuration influences downstream simulation behaviour
- products connect multiple datasets together
- commercial events begin with product configuration
- baseline market share is only the starting point for competition

Understanding the product portfolio is essential before exploring market dynamics.
